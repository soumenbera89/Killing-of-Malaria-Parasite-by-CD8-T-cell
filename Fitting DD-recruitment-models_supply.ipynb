```python
library("readxl")
getwd()
setwd("/home/soumen/Documents/data/separate-data")
my_data = read_excel("para_9.xlsx")
```

```python
my_data1 = data.frame(my_data)
#print(my_data1)
head(my_data1)
my_data1$Tcells[my_data1$Tcells == 0] <- 0.00001
my_data1$Background <- as.numeric(as.character(my_data1$Background))
my_data1$Intensity <- as.numeric(as.character(my_data1$Intensity))
#my_data1$Background[my_data1$Background == NA] <- mean(my_data1$Background,na.rm=TRUE)
my_data1$Background[is.na(my_data1$Background)] <- mean(my_data1$Background,na.rm=TRUE)
my_data1$Intensity[is.na(my_data1$Intensity)] <- mean(my_data1$Intensity,na.rm=TRUE)
my_data1$Time = my_data1$Time/60
print(my_data1)
#mean(my_data1$Background,na.rm=TRUE)
my_data1$Background
```

```python
my_data1$VI <- log10(my_data1$Intensity/my_data1$Background)
my_data1$VI1 <-c(my_data1$VI[2:length(my_data1$VI)],NA)
```

```python
library(minpack.lm)
library(nlmrt)
library(nls2)
library(proto)
# fitting VI data
#expected_VI1<- formula(abs(VI1) ~ P*(abs(VI)^(alpha)))
#expected_VI2<- formula(abs(VI1) ~ P*(abs(VI)^(alpha))*((Tcells)^(beta)))
#expected_VI3<- formula(abs(VI1) ~ P*(abs(VI)^(alpha))*(1-K*(Tcells)))
#expected_VI4<- formula(abs(VI1) ~ P*(abs(VI)^(alpha))*(1-K*((Tcells)^(beta))))
#expected_VI5<- formula(abs(VI1) ~ P*(abs(VI)^(alpha))*(1+K))
#expected_VI6 <- formula(abs(VI1) ~ abs(VI) + P*abs(VI) -K*(Tcells)*(abs(VI)^(alpha)))
#expected_VI7 <- formula(abs(VI1) ~ abs(VI) + P*abs(VI))
DD_model_cluster <- formula(Tcells ~ (A/B)*(exp(B*Time)-1))
```

```python
#fit.VI.2<-nlsLM(expected_VI2, data = my_data1, start=list(P=2.00))
#fit.VI1<-nlsLM(expected_VI1, data = my_data1,start=list(P=0.9,alpha=0.9))
#fit.VI2<-nlsLM(expected_VI2, data = my_data1,start=list(P=8.439e-01,alpha=7.330e-01, beta=-7.206e-02))
#fit.VI3<-nlsLM(expected_VI3,data = my_data1, start=list(P=8.557e-01,alpha=7.330e-01,K=0.02))
#fit.VI4<-nlsLM(expected_VI4,data = my_data1, start=list(P=7.447e-02,alpha=8.123e-01,K=-2.664e+01, beta=-3.759e-01))
#fit.VI5<-nlxb(expectbed_VI5,data = my_data1, start=c(P=0.91,alpha=0.9,K=0.9),trace=TRUE)
#fit.VI6<-nls2(expected_VI5,data = my_data1, start = fit.VI5$coefficients,algorithm = "brute-force")
#fit.VI7 <- nlsLM(expected_VI7, data=my_data1, start = list(P=0.01))
#summary(fit.VI7)
#coef(fit.VI7)
#sum(residuals(fit.VI7)^2)
fit_DD_model <- nlsLM(DD_model_cluster, data = my_data1, start = list(A=1.5,B=0.02),
                      control=list(maxiter=1024,maxfev=10000))
```

```python
summary(fit_DD_model)
#sum(residuals(fit_DD_model)^2)
fit_DD_model
AIC(fit_DD_model)
```
