```python
library("readxl")
library(tidyverse)
library(drc)
getwd()
setwd("/home/soumen/Documents/data/separate-data")
my_data1 = read_excel("Mouse_whole.xlsx")
my_data2 = read_excel("Mouse_t-4-6.xlsx")
my_data3 = read_excel("Mouse_t-6-8.xlsx")
my_data4 = read_excel("Mouse_t-8-10.xlsx")
my_data5 = read_excel("Mouse_t-10-12.xlsx")
my_data6 = read_excel("Mouse_t-8-12.xlsx")
```

```python
#Normalise the data with their mean value
my_data1 <- my_data1 %>% 
mutate(persent_response = abs(VI_Index/(mean(my_data1$VI_Index[my_data1$Tcells==0])))*100)

my_data2 <- my_data2 %>% 
mutate(persent_response = abs(VI_Index/(mean(my_data2$VI_Index[my_data2$Tcells==0])))*100)

my_data3 <- my_data3 %>% 
mutate(persent_response = abs(VI_Index/(mean(my_data3$VI_Index[my_data3$Tcells==0])))*100)

my_data4 <- my_data4 %>% 
mutate(persent_response = abs(VI_Index/(mean(my_data4$VI_Index[my_data4$Tcells==0])))*100)

my_data5 <- my_data5 %>% 
mutate(persent_response = abs(VI_Index/(mean(my_data5$VI_Index[my_data5$Tcells==0])))*100)

my_data6 <- my_data6 %>% 
mutate(persent_response = abs(VI_Index/(mean(my_data6$VI_Index[my_data6$Tcells==0])))*100)
```
